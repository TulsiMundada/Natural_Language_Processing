```python
from nltk import pos_tag , RegexpParser
from nltk.tokenize import word_tokenize
```

```python
sent = 'The Moroccan fairy tale ends at the hands of France.'
```

```python
tokens = word_tokenize(sent)
tags = pos_tag(tokens)
tags
```

```python
chunker = RegexpParser("""
NP: {} 
P: {}                       
V:{}     
PP:{}                       
VP:{<NN><VBZ>}
""" )
result = chunker.parse(tags)
```

```python
result
```

### Dependency Tree

```python
sent = "It was the biggest mistake made by him in the history."
#sent = "He was the biggest criminal , biggest entreprenaur , biggest person of India , he needs to be arrested right now,"
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(sent)
for token in doc:
    print(token.text , '==>' , token.dep_)
```

```python
### Visualize Tree

from spacy import displacy
```

```python
displacy.render(doc)
```

```python
sent = "He is the biggest criminal , he is the biggest bussinessman , he is the biggest national issue."
doc = nlp(sent)
for token in doc:
    print(token.text , '==>' , token.dep_)
displacy.render(doc , jupyter=True)
```
