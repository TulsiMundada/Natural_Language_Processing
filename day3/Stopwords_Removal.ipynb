### punctuation removal 

```python
from nltk.tokenize import word_tokenize
```

```python
sent = 'Hello friends ! How are you ? Welcome to the world of "python Programming".'
print(sent)
```

```python
lst = word_tokenize(sent)
lst
```

```python
clean = [x for x in word_tokenize(sent) if x.isalpha() or x.isdigit()]
' '.join(clean)
```

```python
import string
clean =  [x for x in word_tokenize(sent) if x not in string.punctuation]
' '.join(clean)
```

### Stopwords Removal

```python
from nltk.corpus  import stopwords
swords = stopwords.words('english')
swords
```

```python
sent = 'how are you ? we are students of cdac . I me and myself. How are you ?'
clean = [x for x in word_tokenize(sent) if x.lower() not in swords]
' '.join(clean)
```
