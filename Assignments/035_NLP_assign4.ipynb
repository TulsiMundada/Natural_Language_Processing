# Q.1
- Open the wikipedia page https://en.wikipedia.org/wiki/Saturn
1. Scrap the web and find top 10 nouns from the page
2. Print all the unique verbs in their root form.
3. Extract all the < DT >< JJ >< NN > phrases from the article.
4. Summarize the text using TextRank Algorithm
5. Summarize the text using extractive summarization

```python
import bs4 as bs
import urllib

source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Saturn').read()
soup = bs.BeautifulSoup(source,'lxml')

text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
import string
```

```python
clean_text = [word for word in word_tokenize(text) if word not in string.punctuation]
```

```python
tags = pos_tag(clean_text)
tags
```

```python
nouns = [word[0] for word in tags if word[1].startswith('N')]
nouns
```

```python
from nltk import FreqDist
```

```python
freq = FreqDist(nouns)
freq
```

```python
freq.plot(20)
```

```python
freq.most_common(3)
```

### Print all the unique verbs in their root form.

```python
verbs = [word[0] for word in tags if word[1].startswith('V')]
verbs
```

```python
unique_verbs = list(set(verbs))
unique_verbs
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
for verb in unique_verbs:
    print(verb, "-->", wnl.lemmatize(verb, pos='v'))
```

### Summarize the text using TextRank Algorithm

```python
pip install sumy
```

```python
from sumy.summarizers.text_rank import TextRankSummarizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
my_parser = PlaintextParser(text, Tokenizer('english'))
```

```python
text_rank_summarizer = TextRankSummarizer()
```

```python
summary = text_rank_summarizer(my_parser.document, sentences_count=3)
```

```python
for sent in summary:
    print(sent,'\n')
```

### Summarize the text using Extractive Text Summarization

```python
from sumy.summarizers.lex_rank import LexRankSummarizer
```

```python
# create the object of summarizer
lex_rank_summarizer = LexRankSummarizer()
```

```python
summary = lex_rank_summarizer(my_parser.document, sentences_count=3)
```

```python
for sent in summary:
    print(sent, '\n')
```

```python
# import the class
from sumy.summarizers.lsa import LsaSummarizer

# initialize the parser
lsa_summarizer = LsaSummarizer()

summary = lsa_summarizer(my_parser.document,sentences_count=3)

for sent in summary:
    print(sent,'\n')
```
